```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
b=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=b)
```

```python
a=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in a.containers:
    a.bar_label(bars)
```

```python
b=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=b)
```

```python
b=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=b,x='State',y='Orders')
```

```python
b=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=b,x='State',y='Amount')
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python


b = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)


sns.set(rc={'figure.figsize':(15,5)})


sns.barplot(data=b, x='Marital_Status', y='Amount', hue='Gender')

plt.title("Amount Spent by Marital Status and Gender") 
plt.show()
```
